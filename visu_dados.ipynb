```python
import pandas as pd
import seaborn as srn 
import statistics as sts
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('enhanced_anxiety_dataset.csv')
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
df_occupation= df.groupby(['Occupation']).size()
df_occupation
```

```python
# gr√°fico de barras 
df_occupation.plot.bar()
```

```python
#comparar as profiss√µes com o n√≠vel de ansiedade
group_occupation_anxiety=df.groupby(
    ['Occupation']) ['Anxiety Level (1-10)'].mean()
group_occupation_anxiety.plot.bar()
```

```python
# profiss√µes X Consumo de cafeina
group_occupation_caffeine=df.groupby(
    ['Occupation']) ['Caffeine Intake (mg/day)'].sum()
group_occupation_caffeine.plot.bar()
```

```python
# grafico de pizza
group_occupation_anxiety.plot.pie(autopct = '%1.1f%%')
```

```python
df_gender= df.groupby(['Gender']).size()
df_gender.plot.pie(autopct = '%1.1f%%')
```

```python
## Histrograma 
# dados num√©ricos
# dados agrupados por intervalos de valores (bins)
# distribui√ß√£o de frequencia
```

```python
# histrograma das horas de sono
# kde= linha de densidade
# densidade = melhora a visuali√ß√£o
srn.histplot(df['Sleep Hours'],
            kde= True,
            bins= 15,
            color = 'purple').set_title('Horas de sono')
            
```

```python
# s√≥ a linha de densidade 
srn.kdeplot(df['Sleep Hours'])
```

```python
# histrograma atividade fisica
srn.histplot(df['Physical Activity (hrs/week)'],
            kde= True,
            bins= 10,
            color = 'red').set_title('Atividade F√≠sica')
```

```python
# histrograma comparando dois atributos 
# engenherios X horas de sono
srn.histplot(df.loc[df['Occupation']== 'Engineer']['Sleep Hours'] 
            ).set_title ('Horas de sono dos Engenheiros')
```

```python
# horas de sono dos artistas 
srn.histplot(df.loc[df['Occupation']== 'Artist']['Sleep Hours'] 
            ).set_title ('Horas de sono dos Artistas')
```

```python
# imprimir todos os gr√°ficos em uma figura
# cria uma figura para hospedar os gr√°ficos
plt.figure()

# medicos
plt.subplot(2,2,1)
srn.histplot(
            df.loc[df['Occupation'] == 'Doctor']['Anxiety Level (1-10)'],
             bins = 9
            ).set_title('N√≠vel de ansiedade em m√©dicos')

# enfemeiras
plt.subplot(2,2,2)
srn.histplot(
            df.loc[df['Occupation'] == 'Nurse']['Anxiety Level (1-10)'],
             bins = 9
            ).set_title('N√≠vel de ansiedade em Enfermeiras')

# cientistas
plt.subplot(2,2,3)
srn.histplot(
            df.loc[df['Occupation'] == 'Scientist']['Anxiety Level (1-10)'],
             bins = 9
            ).set_title('N√≠vel de ansiedade em Cientistas')

# m√∫sicos
plt.subplot(2,2,4)
srn.histplot(
            df.loc[df['Occupation'] == 'Musician']['Anxiety Level (1-10)'],
             bins = 9
            ).set_title('N√≠vel de ansiedade em M√∫sicos')

# ajustat o layout para evitar sobreposi√ß√£o
plt.tight_layout()
```
